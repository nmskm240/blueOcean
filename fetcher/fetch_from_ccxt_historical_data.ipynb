## ccxtで仮想通貨のデータをinfluxに追加する

```python
import os

from influxdb_client.client.influxdb_client_async import InfluxDBClientAsync
from blueOcean.ohlcv import OhlcvRepository, CcxtOhlcvFetcher
```

```python
EXCHANGE = 'bybit'  # 例: 'binance', 'bybit', 'bitget'
SYMBOL = "ETH/USDT"
```

```python
async with InfluxDBClientAsync(os.getenv("INFLUXDB_URL"), os.getenv("INFLUXDB_TOKEN"), os.getenv("INFLUXDB_ORG")) as client:
    repository = OhlcvRepository(client)
    fetcher = CcxtOhlcvFetcher(EXCHANGE)
    latest_at = await repository.get_latest_timestamp(EXCHANGE, SYMBOL)
    for batch in fetcher.fetch_ohlcv(SYMBOL, latest_at):
        await repository.save(batch)
```
