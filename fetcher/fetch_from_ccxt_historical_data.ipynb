## ccxtで仮想通貨のデータをinfluxに追加する

```python
from blueOcean.ohlcv import OhlcvRepository, CcxtOhlcvFetcher
```

```python
EXCHANGE = 'bybit'  # 例: 'binance', 'bybit', 'bitget'
SYMBOL = "SOL/USDT"
```

```python
repository = OhlcvRepository("../data")
fetcher = CcxtOhlcvFetcher(EXCHANGE)
latest_at = repository.get_latest_timestamp(EXCHANGE, SYMBOL)
for batch in fetcher.fetch_ohlcv(SYMBOL, latest_at):
    repository.save(batch, EXCHANGE, SYMBOL)
```
